```python
import json
import tensorflow as tf
from tensorflow.keras.layers import (
    Conv2D, Conv2DTranspose, ReLU, GroupNormalization, Softmax, LayerNormalization, DepthwiseConv2D,
    Layer, Dense
)
from tensorflow.keras.models import Model
from tensorflow.keras.activations import tanh, softmax

def pixel_shuffle(fm, r):
    b, h, w, c = fm.shape
    if r == 1:
        return fm
    out_channel = c // (r ** 2)
    out_h = h * r
    out_w = w * r
    fm_view = tf.reshape(fm, (b, h, w, out_channel, r, r))
    fm_prime = tf.transpose(fm_view, perm=[0, 1, 4, 2, 5, 3])
    fm_prime = tf.reshape(fm_prime, (b, out_h, out_w, out_channel))
    return fm_prime

class PixelShuffle(Layer):
    def __init__(self, upscale_factor, dtype='float32', **kwargs):
        super(PixelShuffle, self).__init__(dtype=dtype, **kwargs)
        self.upscale_factor = upscale_factor

    def call(self, inputs):
        return pixel_shuffle(inputs, self.upscale_factor)

    def get_config(self):
        config = super(PixelShuffle, self).get_config()
        config.update({"upscale_factor": self.upscale_factor})
        return config

class CoFusion(Layer):
    def __init__(self, in_ch, out_ch, dtype='float32', **kwargs):
        super(CoFusion, self).__init__(dtype=dtype, **kwargs)
        self.in_ch = in_ch
        self.out_ch = out_ch
        self.conv1 = Conv2D(32, kernel_size=3, strides=1, padding='same', dtype=dtype)
        self.conv3 = Conv2D(out_ch, kernel_size=3, strides=1, padding='same', dtype=dtype)
        self.relu = ReLU(dtype=dtype)
        self.norm_layer1 = GroupNormalization(groups=4, axis=-1, dtype='float32')

    def call(self, x):
        attn = self.relu(self.norm_layer1(self.conv1(x)))
        attn = softmax(self.conv3(attn), axis=-1)
        return tf.expand_dims(tf.reduce_sum(x * attn, axis=-1), axis=-1)

    def get_config(self):
        config = super(CoFusion, self).get_config()
        config.update({"in_ch": self.in_ch, "out_ch": self.out_ch})
        return config

class CoFusion2(Layer):
    def __init__(self, in_ch, out_ch, dtype='float32', **kwargs):
        super(CoFusion2, self).__init__(dtype=dtype, **kwargs)
        self.in_ch = in_ch
        self.out_ch = out_ch
        self.conv1 = Conv2D(32, kernel_size=3, strides=1, padding='same', dtype=dtype)
        self.conv3 = Conv2D(out_ch, kernel_size=3, strides=1, padding='same', dtype=dtype)
        # Заменяем Smish на ReLU
        self.relu = ReLU(dtype=dtype)

    def call(self, x):
        attn = self.conv1(self.relu(x))
        attn = self.conv3(self.relu(attn))
        return tf.expand_dims(tf.reduce_sum(x * attn, axis=-1), axis=-1)

    def get_config(self):
        config = super(CoFusion2, self).get_config()
        config.update({"in_ch": self.in_ch, "out_ch": self.out_ch})
        return config

class DoubleFusion(Layer):
    def __init__(self, in_ch, out_ch, dtype='float32', **kwargs):
        super(DoubleFusion, self).__init__(dtype=dtype, **kwargs)
        self.in_ch = in_ch
        self.out_ch = out_ch
        self.DWconv1 = DepthwiseConv2D(kernel_size=3, strides=1, padding='same', depth_multiplier=8, dtype=dtype)
        self.PSconv1 = PixelShuffle(upscale_factor=1, dtype=dtype)
        self.DWconv2 = DepthwiseConv2D(kernel_size=3, strides=1, padding='same', depth_multiplier=1, dtype=dtype)
        # Заменяем AF (Smish) на ReLU
        self.relu = ReLU(dtype=dtype)

    def call(self, x):
        attn = self.PSconv1(self.DWconv1(self.relu(x)))
        attn2 = self.PSconv1(self.DWconv2(self.relu(attn)))
        # Удаляем Fsmish, так как мы больше не используем эту функцию
        return tf.expand_dims(tf.reduce_sum(attn2 + attn, axis=-1), axis=-1)

    def get_config(self):
        config = super(DoubleFusion, self).get_config()
        config.update({"in_ch": self.in_ch, "out_ch": self.out_ch})
        return config

class DenseLayer(Layer):
    def __init__(self, input_features, out_features, dtype='float32', **kwargs):
        super(DenseLayer, self).__init__(dtype=dtype, **kwargs)
        self.input_features = input_features
        self.out_features = out_features
        self.conv1 = Conv2D(out_features, kernel_size=3, strides=1, padding='same', use_bias=True, dtype=dtype)
        # Заменяем smish1 на ReLU
        self.relu1 = ReLU(dtype=dtype)
        self.conv2 = Conv2D(out_features, kernel_size=3, strides=1, padding='same', use_bias=True, dtype=dtype)

    def call(self, x):
        x1, x2 = x
        # Удаляем Fsmish, так как мы больше не используем эту функцию
        new_features = self.conv2(self.relu1(self.conv1(x1)))
        return 0.5 * (new_features + x2), x2

    def get_config(self):
        config = super(DenseLayer, self).get_config()
        config.update({"input_features": self.input_features, "out_features": self.out_features})
        return config

class DenseBlock(Layer):
    def __init__(self, num_layers, input_features, out_features, dtype='float32', **kwargs):
        super(DenseBlock, self).__init__(dtype=dtype, **kwargs)
        self.num_layers = num_layers
        self.input_features = input_features
        self.out_features = out_features
        self.layers = [DenseLayer(input_features, out_features, dtype=dtype) for _ in range(num_layers)]

    def call(self, x):
        for layer in self.layers:
            x = layer(x)
        return x

    def get_config(self):
        config = super(DenseBlock, self).get_config()
        config.update({
            "num_layers": self.num_layers,
            "input_features": self.input_features,
            "out_features": self.out_features
        })
        return config

class UpConvBlock(Layer):
    def __init__(self, in_features, up_scale, dtype='float32', **kwargs):
        super(UpConvBlock, self).__init__(dtype=dtype, **kwargs)
        self.in_features = in_features
        self.up_scale = up_scale
        self.constant_features = 16
        self.features = self.make_deconv_layers()

    def make_deconv_layers(self):
        layers = []
        for i in range(self.up_scale):
            out_features = 1 if i == self.up_scale - 1 else self.constant_features
            layers.extend([
                Conv2D(out_features, kernel_size=1, dtype=self.dtype),
                ReLU(dtype=self.dtype),
                Conv2DTranspose(out_features, kernel_size=4, strides=2, padding='same', dtype=self.dtype)
            ])
        return tf.keras.Sequential(layers)

    def call(self, x):
        return self.features(x)

    def get_config(self):
        config = super(UpConvBlock, self).get_config()
        config.update({"in_features": self.in_features, "up_scale": self.up_scale})
        return config

class SingleConvBlock(Layer):
    def __init__(self, in_features, out_features, stride, use_ac=False, dtype='float32', **kwargs):
        super(SingleConvBlock, self).__init__(dtype=dtype, **kwargs)
        self.in_features = in_features
        self.out_features = out_features
        self.stride = stride
        self.use_ac = use_ac
        self.conv = Conv2D(out_features, kernel_size=1, strides=stride, use_bias=True, dtype=dtype)
        if self.use_ac:
            self.relu = ReLU(dtype=dtype)

    def call(self, x):
        x = self.conv(x)
        if self.use_ac:
            return self.relu(x)
        else:
            return x

    def get_config(self):
        config = super(SingleConvBlock, self).get_config()
        config.update({
            "in_features": self.in_features,
            "out_features": self.out_features,
            "stride": self.stride,
            "use_ac": self.use_ac
        })
        return config

class DoubleConvBlock(Layer):
    def __init__(self, in_features, mid_features, out_features=None, stride=1, use_act=True, dtype='float32', **kwargs):
        super(DoubleConvBlock, self).__init__(dtype=dtype, **kwargs)
        self.in_features = in_features
        self.mid_features = mid_features
        self.out_features = out_features if out_features is not None else mid_features
        self.stride = stride
        self.use_act = use_act
        self.conv1 = Conv2D(mid_features, kernel_size=3, padding='same', strides=stride, dtype=dtype)
        self.conv2 = Conv2D(self.out_features, kernel_size=3, padding='same', dtype=dtype)
        self.relu = ReLU(dtype=dtype)

    def call(self, x):
        x = self.conv1(x)
        x = self.relu(x)
        x = self.conv2(x)
        if self.use_act:
            x = self.relu(x)
        return x

    def get_config(self):
        config = super(DoubleConvBlock, self).get_config()
        config.update({
            "in_features": self.in_features,
            "mid_features": self.mid_features,
            "out_features": self.out_features,
            "stride": self.stride,
            "use_act": self.use_act
        })
        return config

class TED(Model):
    def __init__(self, dtype='float32', **kwargs):
        super(TED, self).__init__(dtype=dtype, **kwargs)

        self.block_1 = DoubleConvBlock(3, 16, 16, stride=2, dtype=dtype)
        self.block_2 = DoubleConvBlock(16, 32, use_act=False, dtype=dtype)
        self.dblock_3 = DenseBlock(1, 32, 48, dtype=dtype)

        self.maxpool = tf.keras.layers.MaxPool2D(pool_size=3, strides=2, padding='same', dtype=dtype)

        self.side_1 = SingleConvBlock(16, 32, 2, dtype=dtype)
        self.pre_dense_3 = SingleConvBlock(32, 48, 1, dtype=dtype)

        self.up_block_1 = UpConvBlock(16, 1, dtype=dtype)
        self.up_block_2 = UpConvBlock(32, 1, dtype=dtype)
        self.up_block_3 = UpConvBlock(48, 2, dtype=dtype)

        self.block_cat = DoubleFusion(3, 3, dtype=dtype)

    def call(self, x):
        x = tf.cast(x, self.dtype)

        block_1 = self.block_1(x)
        block_1_side = self.side_1(block_1)

        block_2 = self.block_2(block_1)
        block_2_down = self.maxpool(block_2)
        block_2_add = block_2_down + block_1_side

        block_3_pre_dense = self.pre_dense_3(block_2_down)
        block_3, _ = self.dblock_3(([block_2_add, block_3_pre_dense]))

        out_1 = self.up_block_1(block_1)
        out_2 = self.up_block_2(block_2)
        out_3 = self.up_block_3(block_3)

        results = [out_1, out_2, out_3]

        block_cat = tf.concat(results, axis=-1)  # channel-wise concat
        block_cat = self.block_cat(block_cat)

        results.append(block_cat)

        return results

    def get_config(self):
        return super(TED, self).get_config()

    @tf.function(input_signature=[tf.TensorSpec(shape=[None, None, None, 3], dtype=tf.float32)])
    def serve(self, x):
        return self(x)

    def save(self, filepath):
        tf.saved_model.save(self, filepath, signatures={'serving_default': self.serve})

    @classmethod
    def load(cls, filepath):
        loaded = tf.saved_model.load(filepath)
        return loaded.signatures.get('serving_default')
```

```python
# Создание оптимизированной модели (OptimizedTED)
class OptimizedTED(Model):
    def __init__(self, dtype='float32', **kwargs):
        super(OptimizedTED, self).__init__(dtype=dtype, **kwargs)

        # Уменьшение количества фильтров и использование DepthwiseSeparableConv
        self.block_1 = DepthwiseSeparableConv(3, 12, stride=2, dtype=dtype)
        self.block_2 = DepthwiseSeparableConv(12, 24, dtype=dtype)
        self.block_3 = DepthwiseSeparableConv(24, 32, dtype=dtype)

        self.maxpool = tf.keras.layers.MaxPool2D(pool_size=2, strides=2, padding='same', dtype=dtype)

        self.side_1 = tf.keras.layers.Conv2D(24, kernel_size=1, strides=2, dtype=dtype)

        self.up_block_1 = OptimizedUpConvBlock(12, 1, out_channels=1, dtype=dtype)
        self.up_block_2 = OptimizedUpConvBlock(24, 1, out_channels=1, dtype=dtype)
        self.up_block_3 = OptimizedUpConvBlock(32, 2, out_channels=1, dtype=dtype)

        self.final_conv = tf.keras.layers.Conv2D(1, kernel_size=3, padding='same', dtype=dtype)

    def call(self, x):
        x = tf.cast(x, self.dtype)

        block_1 = self.block_1(x)
        block_1_side = self.side_1(block_1)

        block_2 = self.block_2(block_1)
        block_2_down = self.maxpool(block_2)
        block_2_add = block_2_down + block_1_side

        block_3 = self.block_3(block_2_add)

        out_1 = self.up_block_1(block_1)
        out_2 = self.up_block_2(block_2)
        out_3 = self.up_block_3(block_3)

        block_cat = tf.concat([out_1, out_2, out_3], axis=-1)
        final_output = self.final_conv(block_cat)

        return [out_1, out_2, out_3, final_output]

    def get_config(self):
        return super(OptimizedTED, self).get_config()

    @tf.function(input_signature=[tf.TensorSpec(shape=[None, None, None, 3], dtype=tf.float32)])
    def serve(self, x):
        return self(x)

    def save(self, filepath):
        tf.saved_model.save(self, filepath, signatures={'serving_default': self.serve})

    @classmethod
    def load(cls, filepath):
        loaded = tf.saved_model.load(filepath)
        return loaded.signatures.get('serving_default')
```

```python
import tensorflow as tf
import numpy as np
import os
import cv2
from scipy.io import loadmat

class BSDS500Dataset(tf.keras.utils.Sequence):
    def __init__(self, data_root, img_height, img_width, split='train', batch_size=8):
        self.data_root = data_root
        self.img_height = img_height
        self.img_width = img_width
        self.split = split
        self.batch_size = batch_size
        self.mean_bgr = np.array([104.007, 116.669, 122.679])
        self.image_dir = os.path.join(data_root, 'images', split)
        self.gt_dir = os.path.join(data_root, 'groundTruth', split)

        # Create pair_list to store valid image-label pairs
        self.pair_list = self.create_pair_list()

    def __len__(self):
        return len(self.pair_list) // self.batch_size

    def __getitem__(self, idx):
        batch_pairs = self.pair_list[idx * self.batch_size : (idx + 1) * self.batch_size]

        # Check if the batch is empty or too small
        if len(batch_pairs) < self.batch_size:
            # If the batch is too small, pad it with the first sample to reach the desired batch size
            while len(batch_pairs) < self.batch_size:
                batch_pairs.append(self.pair_list[0])  # Pad with the first sample

        batch_images = []
        batch_labels = []

        for img_file, gt_file in batch_pairs:
            img_path = os.path.join(self.image_dir, img_file)
            img = cv2.imread(img_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            gt_path = os.path.join(self.gt_dir, gt_file)
            gt = self.load_gt(gt_path)

            img, gt = self.preprocess(img, gt)
            batch_images.append(img)
            batch_labels.append(gt)
            # Ensure the output shapes are always (8, 352, 352, 3) and (8, 352, 352, 1)

        batch_images = np.array(batch_images)[:self.batch_size]  # Truncate or pad if needed
        batch_labels = np.array(batch_labels)[:self.batch_size]

        return batch_images, batch_labels

    def create_pair_list(self):
        """Creates a list of valid (image_file, ground_truth_file) pairs."""
        image_files = sorted([f for f in os.listdir(self.image_dir) if f.endswith('.jpg')])
        gt_files = sorted([f for f in os.listdir(self.gt_dir) if f.endswith('.mat')])

        pair_list = []
        for img_file in image_files:
            gt_file = os.path.splitext(img_file)[0] + '.mat'
            if gt_file in gt_files:
                pair_list.append((img_file, gt_file))

        return pair_list

    def load_gt(self, gt_path):
        gt_data = loadmat(gt_path)
        gt = gt_data['groundTruth'][0][0][0][0][1]
        return gt

    def preprocess(self, img, gt):
        img = cv2.resize(img, (self.img_width, self.img_height))
        gt = cv2.resize(gt, (self.img_width, self.img_height), interpolation=cv2.INTER_NEAREST)

        img = img.astype(np.float32)
        # img -= self.mean_bgr

        gt = (gt > 0).astype(np.float32)

        if self.split == 'train':
            img, gt = self.augment(img, gt)

        # Keep image in HWC format
        gt = np.expand_dims(gt, axis=-1)  # HW to HWC
        return img / img.max(), gt

    def augment(self, img, gt):
        if np.random.rand() > 0.5:
            img = np.fliplr(img)
            gt = np.fliplr(gt)

        if np.random.rand() > 0.5:
            img = np.flipud(img)
            gt = np.flipud(gt)

        angle = np.random.randint(-10, 11)
        M = cv2.getRotationMatrix2D((self.img_width // 2, self.img_height // 2), angle, 1)
        img = cv2.warpAffine(img, M, (self.img_width, self.img_height))
        gt = cv2.warpAffine(gt, M, (self.img_width, self.img_height))

        return img, gt

# # Example usage
# if __name__ == "__main__":
#     data_root = "data"
#     dataset = BSDS500Dataset(data_root, img_height=352, img_width=352, split='train', batch_size=16)

#     for i in range(len(dataset)):
#         images, labels = dataset[i]
#         print(f"Batch {i}:")
#         print(f"Images shape: {images.shape}")
#         print(f"Labels shape: {labels.shape}")
#         print(f"Images dtype: {images.dtype}")
#         print(f"Labels dtype: {labels.dtype}")
#         print(f"Images min-max: {images.min()}-{images.max()}")
#         print(f"Labels min-max: {labels.min()}-{labels.max()}")
#         print()
```

# losses

```python
import tensorflow as tf

def resize_and_reduce_predictions(prediction, target_shape):
    """
    Resizes predictions to match target shape and reduces channels if needed.
    """
    resized = tf.image.resize(prediction, target_shape[1:3])

    # Use tf.cond for conditional execution within the graph
    resized = tf.cond(
        tf.equal(tf.shape(resized)[-1], target_shape[-1]),  # Condition
        lambda: resized,  # True branch: keep 'resized' as is
        lambda: tf.reduce_mean(resized, axis=-1, keepdims=True)  # False branch: reduce channels
    )

    return resized

def bdcn_loss(y_true, y_pred, l_weight=1.1):
    """
    BDCN loss function for edge detection.

    Args:
        y_true: Ground truth edges.
        y_pred: Predicted edges.
        l_weight: Loss weight.

    Returns:
        Weighted BDCN loss.
    """
    y_true = tf.cast(y_true, tf.float32)
    mask = tf.cast(y_true > 0.5, tf.float32)

    num_positive = tf.reduce_sum(mask)
    num_negative = tf.reduce_sum(1.0 - mask)

    # Add a check to avoid division by zero
    total = num_positive + num_negative
    total = tf.maximum(total, 1e-10)  # Add a small epsilon to prevent zero

    beta = num_negative / total

    pos_weight = beta / (1 - beta)

    # # Clip logits to avoid negative or zero values
    # y_pred = tf.clip_by_value(y_pred, 1e-10, 1.0 - 1e-10)

    loss = tf.nn.weighted_cross_entropy_with_logits(
        labels=y_true,
        logits=y_pred,
        pos_weight=pos_weight
    )

    return tf.reduce_mean(loss) * l_weight

def bdrloss(prediction, label, radius): #modified
    '''
    The boundary tracing loss that handles the confusing pixels.
    '''
    # Создание фильтра
    filt = tf.ones((2*radius+1, 2*radius+1, 1, 1))
    filt = tf.stop_gradient(filt)
    # Вычисление граничных пикселей
    bdr_pred = prediction * label
    pred_bdr_sum = label * tf.nn.conv2d(bdr_pred, filt, strides=[1, 1, 1, 1], padding='SAME')
    # Создание маски для текстуры
    texture_mask = tf.nn.conv2d(tf.cast(label, tf.float32), filt, strides=[1, 1, 1, 1], padding='SAME')
    mask = tf.cast(texture_mask != 0, tf.float32)
    mask = tf.where(label == 1, tf.zeros_like(mask), mask)
    pred_texture_sum = tf.nn.conv2d(prediction * (1 - label) * mask, filt, strides=[1, 1, 1, 1], padding='SAME')
    # Вычисление softmax карты
    softmax_map = tf.clip_by_value(pred_bdr_sum / (pred_texture_sum + pred_bdr_sum + 1e-10), 1e-10, 1 - 1e-10)
    # Вычисление стоимости
    cost = -label * tf.math.log(softmax_map)
    cost = tf.where(label == 0, tf.zeros_like(cost), cost)
    return tf.reduce_sum(tf.reduce_mean(cost, axis=[1, 2, 3]))

def textureloss(prediction, label, mask_radius): #modified
    '''
    The texture suppression loss that smooths the texture regions.
    '''
    # Create filters
    filt1 = tf.ones((3, 3, 1, 1), dtype=tf.float32)
    filt2 = tf.ones((2 * mask_radius + 1, 2 * mask_radius + 1, 1, 1), dtype=tf.float32)

    # Compute sums using convolution
    pred_sums = tf.nn.conv2d(prediction, filt1, strides=[1, 1, 1, 1], padding='SAME')
    label_sums = tf.nn.conv2d(label, filt2, strides=[1, 1, 1, 1], padding='SAME')

    # Compute mask
    mask = 1 - tf.cast(tf.greater(label_sums, 0), tf.float32)

    # Compute loss
    loss = -tf.math.log(tf.clip_by_value(1 - pred_sums / 9, 1e-10, 1 - 1e-10))
    loss = tf.where(tf.equal(mask, 0), tf.zeros_like(loss), loss)

    return tf.reduce_sum(tf.reduce_mean(loss, axis=[1, 2, 3]))

def cats_loss(prediction, label, l_weight=[0., 0.]):
    """
    CATS loss function combining multiple components.
    """
    prediction = resize_and_reduce_predictions(prediction, tf.shape(label))
    tex_factor, bdr_factor = l_weight
    balanced_w = 1.1
    label = tf.cast(label, tf.float32)
    prediction = tf.cast(prediction, tf.float32)

    mask = tf.identity(label)
    num_positive = tf.reduce_sum(tf.cast(tf.equal(mask, 1), tf.float32))
    num_negative = tf.reduce_sum(tf.cast(tf.equal(mask, 0), tf.float32))
    beta = num_negative / (num_positive + num_negative)
    mask = tf.where(tf.equal(mask, 1), beta, balanced_w * (1 - beta))
    mask = tf.where(tf.equal(mask, 2), 0.0, mask)

    prediction = tf.sigmoid(prediction)
    cost = tf.keras.losses.binary_crossentropy(label, prediction, from_logits=False)

    # Dynamically get the batch size from 'cost'
    batch_size = tf.shape(cost)[0]

    if cost.shape != mask.shape:
        # Reshape 'mask' using the dynamic batch size
        mask = tf.reshape(mask, (batch_size, 352, 352, 1))  # Assuming height and width are fixed at 352

    # print(f"cost shape: {cost.shape}")
    # print(f"mask shape: {mask.shape}")

    cost = tf.math.multiply(cost, mask[..., 0])

    label_w = tf.cast(tf.not_equal(label, 0), tf.float32)
    textcost = textureloss(prediction, label_w, mask_radius=4)
    bdrcost = bdrloss(prediction, label_w, radius=4)

    final_loss = cost + bdr_factor * bdrcost + tex_factor * textcost
    final_loss = tf.reduce_mean(final_loss)

    return final_loss

# # Example usage
# if __name__ == "__main__":
#     # Create some dummy data
#     y_true = tf.random.uniform((1, 352, 352, 1), 0, 2, dtype=tf.int32)
#     y_pred = tf.random.normal((1, 352, 352, 1))
#     y_pred2 = tf.random.normal((1, 352, 352, 4))

#     # Compute BDCN loss
#     bdcn_loss_value = bdcn_loss(y_true, y_pred)
#     print(f"BDCN Loss: {bdcn_loss_value.numpy()}")

#     # Compute CATS loss
#     cats_loss_value = cats_loss(y_pred2, tf.cast(y_true, tf.float32))
#     print(f"CATS Loss: {cats_loss_value.numpy()}")
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.mixed_precision import set_global_policy
import wandb
from wandb.integration.keras import WandbCallback

wandb.login(key='badf4f9fe126789d95d5e3dee8f07092031efd54')

# Initialize wandb
wandb.init(project="edge-detection")

# Гиперпараметры
BATCH_SIZE = 32
IMG_HEIGHT = 352
IMG_WIDTH = 352
EPOCHS = 300
INITIAL_LEARNING_RATE = 1e-4
DATA_ROOT = "data"  # Укажите путь к вашим данным
DTYPE = 'float32'

# Log hyperparameters
wandb.config.update({
    "batch_size": BATCH_SIZE,
    "img_height": IMG_HEIGHT,
    "img_width": IMG_WIDTH,
    "epochs": EPOCHS,
    "initial_learning_rate": INITIAL_LEARNING_RATE,
    "dtype": DTYPE
})

# Создание модели
model = TED(dtype=DTYPE)

# Define the optimizer
optimizer = tf.keras.optimizers.Adam(learning_rate=INITIAL_LEARNING_RATE)

# Define learning rate scheduler (adjust as needed)
lr_scheduler = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', factor=0.1, patience=5, verbose=1
)

# Compile the model
model.compile(optimizer=optimizer, loss=cats_loss)  # Use cats_loss

# Create train and validation datasets
train_dataset = BSDS500Dataset(DATA_ROOT, img_height=IMG_HEIGHT, img_width=IMG_WIDTH,
                               split='train', batch_size=BATCH_SIZE)
val_dataset = BSDS500Dataset(DATA_ROOT, img_height=IMG_HEIGHT, img_width=IMG_WIDTH,
                             split='val', batch_size=BATCH_SIZE)

# Define metrics to track
train_loss_metric = tf.keras.metrics.Mean(name='train_loss')
val_loss_metric = tf.keras.metrics.Mean(name='val_loss')

# Loss weights
l_weight0 = [1.1, 0.7, 1.1, 1.3]
l_weight = [[0.05, 2.], [0.05, 2.], [0.01, 1.], [0.01, 3.]]

# Initialize best validation loss
best_val_loss = float('inf')

# Custom training loop
for epoch in range(EPOCHS):
    print(f"Epoch {epoch + 1}/{EPOCHS}")

    # Training loop
    for step in range(len(train_dataset)):
        x_batch_train, y_batch_train = train_dataset[step]
        with tf.GradientTape() as tape:
            preds_list = model(x_batch_train, training=True)
            preds_list = [tf.cast(pred, dtype=tf.float32) for pred in preds_list]

            # Calculate losses
            loss1 = sum([bdcn_loss(y_batch_train, preds, l_w) for preds, l_w in zip(preds_list, l_weight0)])
            loss2 = cats_loss(preds_list[-1], y_batch_train, l_weight[-1])

            if len(loss2.shape) == 3:
                loss2 = tf.expand_dims(loss2, axis=-1)

            loss_value = loss1 + loss2

        grads = tape.gradient(loss_value, model.trainable_weights)
        grads = [tf.clip_by_value(grad, -5.0, 5.0) for grad in grads]
        optimizer.apply_gradients(zip(grads, model.trainable_weights))

        train_loss_metric.update_state(loss_value)

        # Log training loss to wandb
        wandb.log({"train_loss": loss_value.numpy()}, step=epoch * len(train_dataset) + step)

    # Validation loop
    for step in range(len(val_dataset)):
        x_batch_val, y_batch_val = val_dataset[step]
        val_logits = model(x_batch_val, training=False)
        val_logits = [tf.cast(pred, dtype=tf.float32) for pred in val_logits]

        # Calculate losses
        loss1 = sum([bdcn_loss(y_batch_val, preds, l_w) / 10 for preds, l_w in zip(val_logits, l_weight0)])
        loss2 = cats_loss(val_logits[-1], y_batch_val, l_weight[-1])

        val_loss_value = loss1 + loss2
        val_loss_metric.update_state(val_loss_value)

    # Print epoch-level metrics
    train_loss = train_loss_metric.result()
    val_loss = val_loss_metric.result()
    print(f"Training loss: {train_loss.numpy()}, Validation loss: {val_loss.numpy()}")

    # Log epoch-level metrics to wandb
    wandb.log({
        "epoch": epoch,
        "train_loss_epoch": train_loss.numpy(),
        "val_loss_epoch": val_loss.numpy()
    })

    # Save the best model
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        model.save('best_model')  # Save the model weights

        # Save the entire model to wandb
        wandb.save('best_model')

        print("Saved new best model!")

    # Reset metrics for the next epoch
    train_loss_metric.update_state([])
    val_loss_metric.update_state([])

    # Visualization at the end of each epoch
    if epoch % 5 == 0:  # Visualize every 5 epochs (adjust as needed)
        # Get a sample from the validation dataset
        sample_image, sample_label = val_dataset[0]  # Take the first sample
        sample_image = sample_image[0]  # Take the first image in the batch
        sample_label = sample_label[0]

        # Make prediction on the sample
        prediction = model(np.expand_dims(sample_image, axis=0), training=False)[-1]
        prediction = tf.sigmoid(prediction)  # Apply sigmoid to get probabilities
        prediction = prediction.numpy()[0]  # Convert to numpy and remove batch dimension

        # Plot the original image, ground truth, and prediction
        fig, axes = plt.subplots(1, 4, figsize=(15, 5))
        axes[0].imshow(sample_image)
        axes[0].set_title('Original Image')
        axes[1].imshow(sample_label[..., 0], cmap='gray')
        axes[1].set_title('Ground Truth')
        axes[2].imshow(prediction[..., 0], cmap='gray')
        axes[2].set_title('Prediction')
        axes[3].imshow((prediction[..., 0] > 0.5).astype(int), cmap='gray')
        axes[3].set_title('Binarized Prediction')

        # Log the figure to wandb
        wandb.log({"validation_samples": wandb.Image(plt)})

        plt.close(fig)  # Close the figure to free up memory

# Finish the wandb run
wandb.finish()
```

```python
def preprocess_image(image_path, target_size=(352, 352)):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, target_size)
    img = img.astype(np.float32) / 255.0
    img = np.expand_dims(img, axis=0)
    return img

def create_and_load_model(weights_path):
    # Create a new instance of the TED model
    model = TED()

    # # Build the model with a dummy input to initialize weights
    # dummy_input = tf.random.normal((1, 352, 352, 3))
    # _ = model(dummy_input)

    # Load the weights
    model.load(weights_path)

    return model

def run_inference(model, image):
    predictions = model(image, training=False)
    if isinstance(predictions, list):
        prediction = predictions[-1]
    else:
        prediction = predictions
    prediction = tf.sigmoid(prediction)
    return prediction.numpy()[0, ..., 0]

def visualize_results(original_image, prediction):
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(original_image[0])
    axes[0].set_title('Original Image')
    axes[0].axis('off')
    axes[1].imshow(prediction, cmap='gray')
    axes[1].set_title('Edge Prediction')
    axes[1].axis('off')
    axes[2].imshow((prediction > 0.5).astype(int), cmap='gray')
    axes[2].set_title('Binarized Edge Prediction')
    axes[2].axis('off')
    plt.tight_layout()
    plt.show()
```

```python
input_tensor = signature.inputs[0] 
output_tensor = signature.outputs[0] 

class PrunedModel(tf.Module):
    def __init__(self, loaded_model):
        super(PrunedModel, self).__init__()
        self.loaded_model = loaded_model

    @tf.function(input_signature=[tf.TensorSpec(shape=input_tensor.shape, dtype=input_tensor.dtype, name='x')])
    def __call__(self, x):
        outputs = self.loaded_model(x)
        return outputs[0]

pruned_model = PrunedModel(imported)

tf.saved_model.save(pruned_model, 'pruned_model', signatures={'serving_default': pruned_model.__call__})
```
